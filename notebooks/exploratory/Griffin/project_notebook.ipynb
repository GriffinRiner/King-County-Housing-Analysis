```python
def outliers(column):
    Q1 = df[column].quantile(.25)
    Q3 = df[column].quantile(.75)
    IQR = Q3 - Q1
    lower_lim = Q1 - 1.5 * IQR
    upper_lim = Q3 + 1.5 * IQR
    
    return lower_lim, upper_lim
```

```python
def df_no_outliers(df, column):
    outlier_low = (df[column] < outliers(column)[0])
    outlier_high = (df[column] > outliers(column)[1])
    return df[column][~(outlier_low | outlier_high)]
```

```python
df['sqft_living_no_outliers'] = df_no_outliers(df, 'sqft_living')
```
